# Se instalan e importan librerías

```python
import pandas as pd
import zipfile
import seaborn as sns
from matplotlib import pyplot
```

# Se leen los datos

```python
zf = zipfile.ZipFile('profeco.zip') 
# if you want to see all files inside zip folder
zf.namelist() 
# now read your csv file 
```

```python
#read data in chunks of 1 million rows at a time
chunk = pd.read_csv(zf.open('all_data.csv'),chunksize=10000)
data = pd.concat(chunk)
```

```python
data.head()
```

# Análisis Inicial

```python
data.info()
```

```python
#Se limpia la columna de municipio
data['municipio'] = data['municipio'].str.replace(' ','')
```

```python
#Se convierte precio a numérico
data['precio'] = pd.to_numeric(data.precio, errors='coerce')
```

```python
#se valida la conversión
data.info()
```

# Preguntas

**How many commercial chains are monitored, and therefore, included in this database?**

```python
n_cadenas = data.cadenaComercial.nunique()

print('There is {n} chains monitored'.format(n=n_cadenas))
```

**What are the top 10 monitored products by State?**


```python
top_products_by_state = data.groupby('estado')['producto'].value_counts().groupby(level=0, group_keys=False).head(10).to_frame()
```

For better undestanding, I export it as a csv file

```python
top_products_by_state.to_csv('top_products_by_state.csv')
```

**Which is the commercial chain with the highest number of monitored products?**

```python
products_by_chain = data.groupby('cadenaComercial')['producto'].nunique().sort_values(ascending=False).to_frame().reset_index()
```

```python
cadena = data.groupby('cadenaComercial')['producto'].nunique().sort_values(ascending=False).head(1).index[0]
value = data.groupby('cadenaComercial')['producto'].nunique().sort_values(ascending=False).head(1)

print('The commercial chain with the highest number of monitores products is {cc} with {val} products'.format(cc=cadena, val=value))
```

What are the lessons learned from this exercise?

1. It was a very interesting exercise, especially because it states that 80% of the time in data analysis, it takes to organize the data. 
2. It would be interesting to know the business perspective to manage null values
3. The work would be easier through a data pipeline, and have some other visualization tool to automate the process. I think this is where the work of the data engineer would come in.
4. It is good to do the cleaning process for the given sample, however it is better to work with the total data because there are aspects that the initial sample did not have.


**Can you identify other ways to approach this?**

As I said, make a data pipeline would be ideal to manage data and make it available to users to carry out this type of analysis.

Having a meeting with the business to know the real interest of the analysis could speed up the work and generate greater value to the investigation
